```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.zyxware.com/articles/4344/list-of-fortune-500-companies-and-their-websites#:~:text=List%20of%20Fortune%20500%20companies%20and%20their%20websites,%20%20http%3A%2F%2Fwww.berkshirehathaway.com%20%2045%20more%20rows%20'
```

```python
r=requests.get(url)
r.status_code
```

```python
soup=BeautifulSoup(r.content,'lxml')
```

```python
company, website=[],[]
```

```python
for i in soup.find_all('td')[1::3]:
    company.append(i.text)
company
```

```python
for i in soup.find_all('td')[2::3]:
    website.append(i.a['href'])
website
```

```python
import pandas as pd 
```

```python
df=pd.DataFrame({'Company':company,'Website':website})
```

```python
df.to_excel('data.xlsx')
```
